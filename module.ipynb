## 모듈

쉽게 설명하면 어떤 필요한 것들끼리 부품처럼 잘 만들어진 파일이라고 생각하면 된다.

모듈화 == 필요한 것들끼리 부품처럼 잘 만든 것

파이썬에서는 함수 정의나 클래스 등의 파이썬 문장들을 담고 있는 파일을 모듈이라고 한다.

모듈은 확장자가 .py이다.



위 코드가 모듈이 되는 것이다.

코드의 경로랑 같은 경로에 있거나 파이썬 라이브러리들이 모여있는 폴더에 있어야 사용 가능하다.

```python
import th_module
th_module.price(3) # 3명이서 영화 보러 갔을 때
```

```python
th_module.price_morning(4)
```

```python
th_module.price_solider(5)
```

모듈의 이름이 길 때 **as 별명**을 붙여서 사용할 수도 있다.

이름 대신 사용할 수 있다.

```python
import th_module as mv
mv.price(3)
```

from 모듈명 import *
== 모듈에 있는 모든 것을 사용하겠다 라는 의미

위 구문을 사용하면

```python
from th_module import *
price(3)
price_morning(5)
```

모듈명 상관없이 메소드로만 출력할 수 있다.

from import 구문을 사용하면 내가 필요한 부분만 가져와서 사용할 수도 있다.

```python
from th_module import price, price_morning
price(5)
price_morning(6)
```

만약 price_solider을 사용하려고 하면 에러가 뜬다.

```python
from th_module import price, price_morning
price(5)
price_morning(6)
```

위 구문에서도 import 부분이 너무 길어 별명을 사용하고 싶을 때 **as 별명**을 입력하면 된다.

```python
from th_module import price, price_morning as a
a(5)
a(6)
```

## 패키지

패키지는 모듈들을 모아놓은 집합이라고 생각하면 된다.

하나의 디렉토리에 여러 모듈 파일들을 가져다 놓은 것을 패키지라고 쉽게 이해하면 된다.

```python
import introduction.name #디렉토리명 뒤에는 모듈이나 패키지만 가능, 클래스나 함수는 import하지 못한다.
intro = introduction.name.namePackage()
intro.detail()
```

```python
import introduction.name.namePackage
intro = introduction.name.namePackage()
intro.detail()
```

패키지나 함수를 가져다가 import하면 위처럼 에러를 발생하게 된다.

하지만 from ~ import 구문에서는 사용이 가능하다.

```python
from introduction.name import namePackage
#introduction 디렉토리에 있는 name이라는 모듈에서 namePackage를 import한다.
intro = namePackage()
intro.detail()
```

```python
from introduction import age
intro = age.agePackage()
intro.detail()
```

```python
from introduction import * 
intro = age.agePackage()
intro.detail()

#원래는 오류가 뜸
```

from introduction import * 는 introduction 디렉토리에 있는 모든 것을 가져온다는 것인데 오류가 나는 이유는 개발자가 공개 범위를 설정해줘야 된다.



그래서 __init__.py에 가서 보여줄 모듈을 입력해줘야 된다.

입력했으니 실행될 것이다.

```python
from introduction import * 
#intro = age.agePackage()
intro = name.namePackage()
intro.detail()

```

공개 범위에 없는 것은 위와 같이 에러가 뜬다.

## 외장함수

직접 import를 해줘야 쓸 수 있는 함수<br>

https://docs.python.org/3.8/py-modindex.html#cap-r<br>

버전별로 쓸 수 있는 외장함수에 대해서 잘 나와있다.

```python
#glob : 경로 내의 폴더 / 파일 목록 조회 (윈도우 dir과 같다.)

import glob
print(glob.glob("*.py")) # 확장자가 py인 모든 파일
```

```python
# os : 운영체제에서 제공하는 기본 기능

import os
print(os.getcwd()) #현재 디렉토리

folder = "sample_dir"

if os.path.exists(folder):
    print("이미 존재하는 폴더입니다.")
    os.rmdir(folder)
    print("폴더를 삭제했습니다.")
else:
    os.makedirs(folder) # 폴더 생성
    print(folder, "폴더를 생성하였습니다.")
```

```python
print(os.listdir()) # glob과 비슷하다.
```

```python
# time : 시간 관련 함수 제공

import time
print(time.localtime())
print(time.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
import datetime
print("오늘 날짜는 ", datetime.date.today())
```

```python
#timedelta : 두 날짜 사이의 간격

today = datetime.date.today()
td = datetime.timedelta(days=100)
print("우리가 만난지 100일은", today + td)
```
